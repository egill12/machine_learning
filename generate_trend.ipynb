```python
import numpy as np
#np.random.seed(1)
n_samples = 1000
b = 0.2 #  ( range from -0.2 to 0.2 to move from mean reversion to strong trend
a = (1/(1-(b**3)))
x = w = np.random.normal(size=n_samples)*0.001
for t in range(n_samples):
    x[t] = a*x[t-1] + w[t]
    
from matplotlib import pyplot as plt
plt.plot(x)
plt.show()
```
